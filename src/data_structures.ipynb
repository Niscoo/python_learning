#### Les sets

```python
# si on utilise un set c'est qu'on n'a pas besoin de l'ordre ni des répétition sinon nous devons utiliser les listes
# les ops mathématiques comme intersection et union ne sont possibles qu'en utilisant les sets
# dans les sets on peut ajouter des éléments mes pas modifier ceux déjà présents

emails = {"alice@email.com", "bob@email.com", "alice@email.com"}
print(emails)  # {'alice@email.com', 'bob@email.com'} (doublon supprimé)
print("alice@email.com" in emails)  # True (O(1) rapide !)
emails.add("claire@email.com")
emails.add("alice@email.com") # tentative d'ajout d'un doublon, ça ne fait rien du tout

print(emails)  
```

    Exo 4 (dict) : tu as stocks = {"pomme": 10, "banane": 5, "orange": 8} et ventes = [("pomme", 3), ("banane", 2), ("pomme", 4)]. Mets à jour stocks en fonction des ventes (si stock < vente, ne descends pas en-dessous de 0).


```python
l1 = [1, 2, 3, 4]
l2 = [4, 5, 6]

print("les éléments communs:", [x for x in l1 if x in l2])

print("les éléments qui sont seulement dans l1", [x for x in l1 if x not in l2])

print("les éléments présents dans l1 ou l2", list(set(l1) | set(l2)))

test = list(set(l1) | set(l2))
print(test)
```

```python
stocks = {"pomme": 10, "banane": 5, "orange": 8}
ventes = [("pomme", 3),("banane", 2), ("pomme", 10), ("kiwi", 1)]


# for f, q in ventes: 
#     stocks[f] = max(0, stocks[f] - q)

# on utilise dict.get pour éviter une erreur si le fruit n'existe pas, ça rend 0 si c'est le cas
for fruit, q in ventes: stocks[fruit] = max(0, stocks.get(fruit, 0) - q)

print(stocks)
```

### 3. Niveau difficile

    Exo 5 (combination) : à partir d’une liste de tuples
    achats = [("Alice", "pomme"), ("Bob", "banane"), ("Alice", "banane"), ("Clara", "pomme")]
    construis un dict clients → set de produits achetés.
    Exemple attendu : {"Alice": {"pomme", "banane"}, ...}.

    Exo 6 (dict + list) : tu as un dict notes = {"Alice": [12, 15], "Bob": [9, 10, 8], "Clara": [18]}.
    Calcule un nouveau dict moyennes = {"Alice": 13.5, ...} et affiche le nom de l’élève avec la meilleure moyenne.


```python
achats = [("Alice", "pomme"), ("Bob", "banane"), ("Alice", "banane"), ("Clara", "pomme"), ("Alice", "banane")]

dictionnaire = {}



for person, fruit in achats:
    if person not in dictionnaire:
        dictionnaire[person] = set()
    dictionnaire[person].add(fruit)
print(dictionnaire)

dictionnaire = {person: {fruit for p, fruit in achats if p == person} for person, _ in achats}
print(dictionnaire)

```

```python
notes = {"Alice": [12, 15], "Bob": [9, 10, 8], "Clara": [18]}
#notes.values renvoie [[12, 15], [9, 10, 8], [10, 7, 12]] le max compare les liste premier avec premier si egale => deuxième...
print(max(notes.values()))

# dictionnaire pour la moyenne de chaque élève
moyennes = {name: sum(values)/len(values) for name, values in notes.items() }

print(moyennes)
# max moyennes renvoie la personne (l'index) qui a la plus haute moyenne
print(max(moyennes))

# moyennes.values renvoie une liste des moyennes de tous les élèves, max renvoie la meilleure
print(max(moyennes.values()))

print(moyennes.get("Alice"))

meilleur = max(moyennes, key= moyennes.get)
print (meilleur, moyennes[meilleur])
```

### 4. Niveau extra difficile (type entraînement DSA)

    Exo 7 (analyse de fréquence) : écris une fonction qui prend une chaîne, par ex "abracadabra", et retourne :

        un dict caractère → nombre d’apparitions

        le caractère le plus fréquent (en cas d’égalité, n’importe lequel des max).

    Exo 8 (déduplication + ordre) : on te donne une liste de logs d’accès :
    logs = ["alice", "bob", "alice", "clara", "bob", "bob", "dave"]

        affiche la liste des utilisateurs uniques dans l’ordre d’apparition

        affiche un dict utilisateur → nombre d’accès.


```python
def dict_carac_occu (chaine):

    caracteres = {}
    caracteres = {car: chaine.count(car) for car in chaine if car not in caracteres } 

    return caracteres

diction = dict_carac_occu("banana")
print(diction)

print("le caractère le plus fréquent est", max(diction, key= diction.get))
```

```python
logs = ["alice", "bob", "alice", "clara", "bob", "bob", "dave"]
list_person = dict.fromkeys(logs)
print(list_person)

for idx in list_person: list_person[idx] = logs.count(idx) 
print(list_person)

```

#### Given two lists, l1 and l2, write a program to create a third list l3 by picking an odd-index element from the list l1 and even index elements from the list l2.

```python
l1 = [3, 6, 9, 12, 15, 18, 21]
l2 = [4, 8, 12, 16, 20, 24, 28]
l3 = []

for i in range(len(l1)): 
    if i%2 == 1 : l3.append(l1[i])


for i in range(len(l2)):
    if i%2 == 0 : l3.append(l2[i])
    
# l3.sort()
print(l3)
```

#### Slice list into 3 equal chunks and reverse each chunk

```python
sample_list = [11, 45, 8, 23, 14, 12, 78, 45, 89]
nb_elem = len(sample_list)/3
l1 = []
l2 = []
l3 = []

for i in range(int(nb_elem)):
    l1.append(sample_list[i])
    l2.append(sample_list[i+3])
    l3.append(sample_list[i+6])

l1.reverse()
l2.reverse()
l3.reverse()

print(l1, l2, l3)
```

#### Write a code to find the intersection (common) of two sets and remove those elements from the first set.

```python
first_set = {23, 42, 65, 57, 78, 83, 29}
second_set = {57, 83, 29, 67, 73, 43, 48}

intersection = first_set.intersection(second_set)
print(f"Intersection is {intersection}")

for value in intersection:
    first_set.remove(value) 

first_set.add(65)
print(f"First Set after removing common element {first_set}")
```

#### Write a code to get all values from the dictionary and add them to a list but don’t add duplicates

```python
speed = {'jan': 47, 'feb': 52, 'march': 47, 'April': 44, 'May': 52, 'June': 53, 'july': 54, 'Aug': 44, 'Sept': 54}

# order not preserved
my_list = list(set(speed.values()))
print(my_list)

# to preserve order
my_list = []
dict_values = speed.values()
for value in dict_values:
    if value not in my_list: my_list.append(value) 
print(my_list)
```
